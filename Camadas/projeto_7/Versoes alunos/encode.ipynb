```python
from suaBibSignal import *
import numpy as np
import sounddevice as sd
import matplotlib.pyplot as plt
import time
import math
```

```python
#Definindo parametros
sample_rate = 48000 #sample rate da placa de som
sinal = signalMeu() #instancia a classe
tabela_frequencia = [[941, 1336], [697, 1209], [697, 1336], [697, 1477], [770, 1209], [770, 1336], [770, 1477], [852, 1209], [852, 1336], [852, 1477]] #tabela de frequencias (0 - 9)
periodo = 1/sample_rate #periodo de amostragem
amplitude = 1 #amplitude da senoide
duracao = 5 #duração do som em segundos
```

```python
numero_para_som = int(input("Digite um número de 0 a 9: ")) #recebe o número a ser codificado
print(f'gerando som para o número {numero_para_som}')

# equação para a senoide: f(t) = A*sin(2*pi*f*t)
tempo = np.arange(0, duracao, periodo) #cria o vetor tempo
senoide1 = amplitude * np.sin(2*np.pi*tabela_frequencia[numero_para_som][0]*tempo) #cria a primeira senoide
senoide2 = amplitude * np.sin(2*np.pi*tabela_frequencia[numero_para_som][1]*tempo) #cria a segunda senoide
soma_senoide = senoide1 + senoide2 #soma as senoides

print('Executando as senoides (emitindo o som)')
sd.play(soma_senoide, sample_rate) #emite o som
sd.wait() #aguarda o fim do som
```

```python
sinal.plotFFT(soma_senoide, sample_rate)
plt.xlim(500, 2000)
plt.show()

plt.plot(tempo, soma_senoide)
plt.title('tempo por soma das senoides')
plt.xlabel('tempo')
plt.ylabel('senoides somadas')
plt.xlim(0, 0.02)
plt.show()
```
